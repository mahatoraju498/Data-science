```python
import numpy as np
```

```python
np.random.seed(0)
sample_weight = np.random.randint(low = 46, high = 55, size = 25)
```

```python
sample_weight
```

```python
H0 = "There is no diference in means"
H1 = "There is significant difference in mean" # language miluane
```

```python
alpha = 0.05
```

```python
#t test:we want to compare the mean of referance and sample
from scipy import stats
result = stats.ttest_1samp(sample_weight,popmean =mu)
result
```

```python
if result.pvalue <alpha:
  print("we rejwct H0")
else:
    print("we fail to reject H0")
```
