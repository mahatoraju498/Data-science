# Kurtosis refers to the degree of presence of outliers in the distribution. Kurtosis is a statistical measure, whether the data is heavy-tailed or light-tailed in a normal distribution.


#There are 3 types of Kurtosis:
* 1. Mesokurtic
* 2. Platykurtic
* 3. Leptokurtic

#Mespkurtic
* Normal Distribution : mean = $\mu$ ; variance = $\sigma^2$

* Medium Outlier
* Kurtosis = 3

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
np.random.seed (0)
data = np.random .normal ( loc = 5,scale =1 , size = 1500) #mean =
kurtosis= kurtosis(data,fisher = False)#kurtosis can be calculated using either using fisher method  or person method.fisher = false means person method
plt.title(f"kurtosis ={round(kurtosis,2)}")
sns.histplot(data, kde = True)
plt.show()
```

#Platykurtic
* Uniform distributio
* No outliers
* Kurtosis <3


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
np.random.seed (0)
data = np.random.uniform(low =3, high = 6,size = 15000)#yo matra change
kurtosis = kurtosis(data , fisher = False)
plt.title(f"kurtosis = {round(kurtosis,2)}")
sns.histplot(data, bins = 3)
plt.show()
```

#Leptokurtic
* Laplace distribution
* High  Outliers
* Kurtosis >3

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
np.random.seed (0)
data = np.random.laplace( loc = 5,scale =1 , size = 1500) #mean =
kurtosis= kurtosis(data,fisher = False)#kurtosis can be calculated using either using fisher method  or person method.fisher = false means person method
plt.title(f"kurtosis ={round(kurtosis,2)}")
sns.histplot(data, kde = True)
plt.show()
```
