#Line Chart

```python
month = ['jan','feb','march','apr','may','jun']
growth = [3,3.6,2,2.3,-2,3]
```

```python
plt.plot(month, growth)
```

```python
plt.plot(month, growth,'o-')
```

```python
plt.plot(month, growth ,'o-',color = 'purple')
plt.xlabel('month')
plt.ylabel('Growth Percentage (%)')
plt.show()
```

```python
month = [ 'jan','feb','march','apr','may','jun']
actual_growth = [2,3,3.5,4,4.5,6]
predicted_growth=[3,3.5,5,5.5,7,7.5]
plt.plot(month,actual_growth,'o-',label='actual')
plt.plot(month, predicted_growth,'o-',label = 'Predicated')
plt.xlabel('month')
plt.ylabel('Growth percentage(%)')
plt.title('Acutal vs Predicated Growth')
plt.legend()#plots the label in plt.plot()
plt.show()
```

#The relationship between two variables is shown by correlation and we can visualize using it heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
df= sns.load_dataset('iris')
```

```python
df.head(3)
```

```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only=True),annot = True)
```

* The correlation between petal_length and petal_width is 0.96.It is a positive correlation.It  means when the value of petal_length increasea the value of petal_width also increase.you can visulize using scatterplot

```python
sns.scatterplot(x='petal_length',y = 'petal_width',data = df)
```

```python
import numpy as np
x = np.linspace(-5,5,100)
sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)
```

```python
fig, axes =plt.subplots(nrows = 2 , ncols = 2, figsize = (10,5))
ax = axes[0,0]
ax.plot(x,sinx,label = 'sinx')
ax.legend()
ax.set_xlabel('x')
```

```python
fig, axes =plt.subplots(nrows = 2 , ncols = 2, figsize = (10,5))
ax = axes[0,0]
ax.plot(x,sinx,label = 'sinx')
ax.legend()


ax = axes[0,1]
ax.plot(x,cosx,label = 'cosx')
ax.legend()
```

```python
fig ,axes =plt.subplots (nrows = 2, ncols = 2 , figsize = (10,5))
ax = axes[1,0]
ax.plot(x,tanx,label = 'tanx')
ax.legend()
```
