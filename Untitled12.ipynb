```python
import pandas as pd
```

```python
employee = pd.read_excel("/content/employee-dept-table.xlsx", sheet_name = 'employee')
department = pd.read_excel("/content/employee-dept-table.xlsx", sheet_name = 'dept')
```

```python
department
```

#Inner Join

```python
pd.merge(employee,department,on = 'DepartmentID',how = 'inner')
```

#full outer join

```python
pd.merge(employee,department,on = 'DepartmentID',how = 'outer')
```

#left join

```python
pd.merge(employee,department,on = 'DepartmentID', how = 'left')
```

```python
#NaN means missing value ; the departmentID 106 has no department name and location so value is missing
```

```python
pd.merge(employee,department, on = 'DepartmentID', how = 'right')
```
