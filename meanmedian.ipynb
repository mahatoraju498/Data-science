```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
image= cv.imread('/content/mydrive/My Drive/meanmedian.png')
image=np.flip(image, axis=-1)
```

```python
plt.imshow(image)
```

```python
Median=image
prop =Median.shape
```

```python
for i in range(1, prop[0]-2):
  for j in range(1, prop[1]-2):
    win = []
    for x in range(i-1,i+2):
      for y in range(j-1,j+2):
        win.append(Median[x][y])
    Median[i][j] = np.min(win)
```

```python
plt.imshow(Median)
```
